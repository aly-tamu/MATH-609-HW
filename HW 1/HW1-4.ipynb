```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Thomas_alg(n):
    # Givens
    N = 2**n - 1
    h = 2**(-2)
    eps = 1e-3
    f = np.zeros([N,1])
    for i in range(0,N):
        f[i] = 2*i*h + 1
    A_mid = (2*eps + h**2)/(h**2)
    A_tri = -eps/(h**2)
    
    # Thomas algorithm
    # Ly=f
    y = np.zeros([N,1])
    u = np.zeros([N,1])
    y[0] = f[0]
    u[0] = A_mid
    for k in range(1,N):
        lk = A_tri/u[k-1]
        y[k] = f[k] - lk*y[k-1]
        u[k] = A_mid - lk*A_tri
    
    # Ux=y
    x = np.zeros([N,1])
    x[-1] =  y[-1]/u[-1]
    for k in range(N-2,0,-1):
        x[k] = (y[k] - A_tri*x[k+1])/u[k]
        
    return x
```

```python
for n in range(1,8+1):
    ans = Thomas_alg(n)
    plt.figure()
    plt.plot( np.linspace(1,ans.shape[0],ans.shape[0])*h ,ans, marker='o')
    plt.title('Tridiagonal Matrix Solution : n = ' + str(8))
```

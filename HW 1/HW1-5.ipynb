```python
A1 = np.zeros([4,4])
for i in range(0,A1.shape[0]):
    for j in range(0,A1.shape[1]):
        if i == j:
            A1[i,j] = 2
        if (j == i-1) or (j == i+1):
            A1[i,j] = -1
print(A1)

A2 = np.zeros([4,4])
for i in range(0,A2.shape[0]):
    for j in range(0,A2.shape[1]):
        A2[i,j] = 1/(i+j+1)
print(A2)
```

```python
def Cholesky_alg(A):
    # Givens
    N = A.shape[0]
    l = np.zeros(A.shape)
    u = np.zeros(A.shape)
    
    for i in range(0,N):
        for j in range(0,i+1):
            
            summ = 0
            for k in range(0,j):
                summ += l[i,k]*l[j,k]

            if i==j:
                l[i,j] = np.sqrt(A[i,i] - summ)
            else:
                l[i,j] = (1/l[j,j])*(A[i,j] - summ)
            
            u = l.transpose()
            
        
    return l, u
```

```python
l1, u1 = Cholesky_alg(A1)
l2, u2 = Cholesky_alg(A2)
```

```python
print('First Matrix Factorization')
print('------------------')
print(l1)
print('------------------')
print(u1)

print('\nSecond Matrix Factorization')
print('------------------')
print(l2)
print('------------------')
print(u2)
```
